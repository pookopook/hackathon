```python
import pandas as pd
import holidays
```

```python
# 날짜 생성
df = pd.DataFrame({'date': pd.date_range("2021-12-25", "2024-1-15")})

# 한국 공휴일
kr_holidays = holidays.KR(years=range(2021,2025))

# 공휴일 또는 주말 → 0/1 처리
df['is_offday'] = (df['date'].isin(kr_holidays) | (df['date'].dt.weekday >= 5)).astype(int)

# 근무일 여부 → 0/1 처리
df['is_workday'] = (1 - df['is_offday'])  # 또는: (~df['is_offday'].astype(bool)).astype(int)

# workday_block 번호 부여
df['workday_block'] = (df['is_workday'] != df['is_workday'].shift()).cumsum()

# workday_streak 계산
df['workday_streak'] = df.groupby('workday_block')['is_workday'].transform(
    lambda x: x.cumsum() if x.iloc[0] == 1 else 0
)
```
